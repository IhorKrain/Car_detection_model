```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Set base directory path
base_dir = "C:/Users/Public/pycharm/Yolo_detection/model_comparing"
model_dirs = {
    "YOLO8n": os.path.join(base_dir, "Yolo8n_metrics"),
    "YOLO8s": os.path.join(base_dir, "Yolo8s_metrics")
}

# List of metrics to compare
metrics = [
    "confusion_matrix_normalized.png",
    "F1_curve.png",
    "P_curve.png",
    "PR_curve.png",
    "R_curve.png"
]

# Create separate figures for each metric pair
for metric in metrics:
    # Create new figure for each metric with larger size
    plt.figure(figsize=(15, 8))
    
    # Load images
    img1 = mpimg.imread(os.path.join(model_dirs["YOLO8n"], metric))
    img2 = mpimg.imread(os.path.join(model_dirs["YOLO8s"], metric))
    
    # Create subplots side by side
    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.title(f"YOLO8n: {metric.split('.')[0]}", fontsize=12)
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.title(f"YOLO8s: {metric.split('.')[0]}", fontsize=12)
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```
