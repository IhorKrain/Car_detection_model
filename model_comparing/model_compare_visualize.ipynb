```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_metric_pair(base_dir, metric_name):
    """
    Displays a pair of graphs for a single metric
    """
    plt.figure(figsize=(18, 8))
    
    # Model paths
    model1_path = os.path.join(base_dir, "Yolo8n_metrics", metric_name)
    model2_path = os.path.join(base_dir, "Yolo8s_metrics", metric_name)
    
    try:
        # Plot YOLO8n
        plt.subplot(1, 2, 1)
        plt.imshow(mpimg.imread(model1_path))
        plt.title("YOLOv8n", fontsize=12)
        plt.axis('off')

        # Plot YOLO8s
        plt.subplot(1, 2, 2)
        plt.imshow(mpimg.imread(model2_path))
        plt.title("YOLOv8s", fontsize=12)
        plt.axis('off')

        plt.tight_layout()
        plt.show()

        
    except FileNotFoundError as e:
        print(f"Ошибка: {e}")
```

```python

show_metric_pair(
    base_dir="C:/Users/Public/pycharm/Yolo_detection/model_comparing",
    metric_name="confusion_matrix_normalized.png"
)
```

YOLOv8s outperforms YOLOv8n in detecting cars, cyclists, and vans but has higher background misclassification. YOLOv8n performs better for detecting seated persons and trucks. If vehicle detection is the priority.

```python
show_metric_pair(
    base_dir="C:/Users/Public/pycharm/Yolo_detection/model_comparing",
    metric_name="F1_curve.png"
)
```

YOLOv8s achieves a slightly higher F1-score (0.83 vs. 0.82) with more stable curves. It improves accuracy for most classes, especially cars and cyclists, while YOLOv8n performs better for trucks and seated persons.

```python

show_metric_pair(
    base_dir="C:/Users/Public/pycharm/Yolo_detection/model_comparing",
    metric_name="P_curve.png"
)
```

YOLOv8n achieves perfect precision at a confidence threshold of 1.0, while YOLOv8s reaches the same at 0.96, indicating slightly better robustness in YOLOv8n. However, YOLOv8s shows a more consistent precision increase across confidence levels, making it more reliable for lower thresholds
